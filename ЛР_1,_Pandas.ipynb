# **Описание используемого набора данных.**
Используемый набор данных содержит информацию о разнообразных аспектах здоровья и питания взрослых и детей в Соединенных Штатах. Каждая строка представляет отдельного респондента, а столбцы содержат следующие переменные:
- **age_group**: Подгруппа респондентов в возрасте 65 лет и старше были определены как “Senior”(пожилые), а все лица в возрасте до 65 лет - как “Adult”(взрослые)
- **RIDAGEYR**: Возраст респондента
- **RIAGENDR**: Пол респондента (мужской('1')/женский('2'))
- **PAQ605**: Респондент еженедельно занимается физическими упражнениями('1'), а ('2') означает, что он этого не делает, ('7') означает, что он изредка делает физиечкие упражнения
- **BMXBMI**: Индекс массы тела респондента
- **LBXGLU**: Уровень глюкозы в крови респондента после голодания
- **DIQ010**: Тип сахарного диабета респондента
- **LBXGLT**: Уровень перорального инсулина в крови респондента
- **LBXIN**: Уровень инсулина в крови респондента

Ссылка на источник: https://archive.ics.uci.edu/dataset/887/national+health+and+nutrition+health+survey+2013-2014+(nhanes)+age+prediction+subset

```python
# Импортируем библиотеки
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Загрузка данных в датафрейм
df = pd.read_csv('NHANES_age_prediction.csv', delimiter=',')
df.head(10)
```

```python
# Общая информация о наборе данных
print(df.info())
```

```python
# Вывод статистической информации о наборе данных
print(df.describe())
```

```python
# Вывод названия строк и столбцов
print('Строки: ', df.index.values)
print('Столбцы: ', df.columns.values.tolist())
```

```python
# Проверка на отсутствующие значения
print(df.isnull().sum())
# Удаление строк с пропущенными значениями, если они есть
df.dropna(inplace=True)
```

```python
# Заменяем категориальные данные количественными
df = pd.get_dummies(df)
df
```

```python
# Тепловая карта корреляционной матрицы
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

# Визуализации

```python
plt.figure(figsize=(12, 6))
# Визуализация по возрасту
plt.subplot(1, 2, 1)
sns.histplot(df['RIDAGEYR'], bins=30, kde=True)
plt.title('Распределение возраста')
plt.xlabel('Возраст')
plt.ylabel('Частота')
plt.tight_layout()
plt.show()
```

Гистограмма показывает, что большинство респондентов находятся в возрасте от 12 до 15 лет.

```python
plt.figure(figsize=(14, 6))

# Гистограмма по полу
plt.subplot(1, 2, 1)
sns.countplot(x='RIAGENDR', data=df)  # 1 - Мужской, 2 - Женский
plt.title('Распределение по полу')
plt.xlabel('Пол (1 - Мужской, 2 - Женский)')
plt.ylabel('Частота')

# Визуализация по полу и возрасту
plt.subplot(1, 2, 2)
sns.countplot(x='age_group_Adult', hue='RIAGENDR', data=df)
plt.title('Возрастные группы по полу')
plt.xlabel('Возрастная группа (True - люди до 65 лет, False - люди в возрасте 65 лет и старше)')
plt.ylabel('Частота')

plt.tight_layout()
plt.show()
```

Количество респондентов женского пола чуть больше, чем мужского. В возрастной группе людей в возрасте 65 лет и старше примерно одинаковое количество респондентов женского и мужского пола. А в группе людей до 65 лет респондентов женского пола чуть больше, чем мужского.

```python
plt.figure(figsize=(14, 6))

# Гистограмма ИМТ
plt.subplot(1, 2, 1)
sns.histplot(df['BMXBMI'], bins=30, kde=True)
plt.title('Распределение ИМТ')
plt.xlabel('Индекс массы тела')
plt.ylabel('Частота')

# Ящик с усами для ИМТ по возрастной группе
plt.subplot(1, 2, 2)
sns.boxplot(x='age_group_Adult', y='BMXBMI', data=df)
plt.title('ИМТ по возрастной группе')
plt.xlabel('Возрастная группа (True - люди до 65 лет, False - люди в возрасте 65 лет и старше)')
plt.ylabel('Индекс массы тела')

plt.tight_layout()
plt.show()
```

Гитограмма показывает, что чаще всего встречаются респонденты с ИМТ = 26-28(более 250 человек). В среднем респонденты из обеих возрастных категорий имеют схожие значения ИМТ. Разброс данных несколько шире для группы респондентов до 65 лет. Среди респондентов до 65 лет встречаются люди с крайне высокими значениями ИМТ (выше 50).

```python
plt.figure(figsize=(12, 6))

# Гистограмма уровня глюкозы
plt.subplot(1, 2, 1)
sns.histplot(df['LBXGLU'], bins=30, kde=True)
plt.title('Распределение уровня глюкозы в крови')
plt.xlabel('Уровень глюкозы')
plt.ylabel('Частота')

# Ящик с усами для уровня глюкозы по полу
plt.subplot(1, 2, 2)
sns.boxplot(x='RIAGENDR', y='LBXGLU', data=df)
plt.title('Уровень глюкозы по полу')
plt.xlabel('Пол (1 - Мужской, 2 - Женский)')
plt.ylabel('Уровень глюкозы')

plt.tight_layout()
plt.show()
```

Гистограмма показывает, что у большинства респондентов уровень глюкозы не превышает 100. Обе группы (мужчины и женщины) имеют схожие медианные значения уровня глюкозы, но среди женщин наблюдается немного больший разброс данных. Наличие большого числа выбросов в обеих группах, особенно на уровне выше 200 и 300, свидетельствует о возможности наличия значительного числа людей с аномальными уровнями глюкозы, что может быть связано с заболеванием, таким как диабет.

```python
plt.figure(figsize=(14, 6))

# Гистограмма уровня инсулина
plt.subplot(1, 2, 1)
sns.histplot(df['LBXIN'], bins=30, kde=True)
plt.title('Распределение уровня инсулина')
plt.xlabel('Уровень инсулина')
plt.ylabel('Частота')

# Ящик с усами для уровня инсулина по возрастной группе
plt.subplot(1, 2, 2)
sns.boxplot(x='age_group_Adult', y='LBXIN', data=df)
plt.title('Уровень инсулина по возрастной группе')
plt.xlabel('Возрастная группа (True - люди до 65 лет, False - люди в возрасте 65 лет и старше)')
plt.ylabel('Уровень инсулина')

plt.tight_layout()
plt.show()
```

Гистограмма показывает, что у большинства респондентов уровень инсулина не превышает 8. Люди до 65 лет демонстрируют более высокий и более вариативный уровень инсулина. У пожилых людей (65+) уровень инсулина в целом ниже и более стабилен. В группе до 65 лет чаще встречаются экстремально высокие значения уровня инсулина.

# Групповые визуализации

```python
# Уровень глюкозы по возрастной группе и полу
plt.figure(figsize=(10, 6))
sns.barplot(x='age_group_Adult', y='LBXGLU', hue='RIAGENDR', data=df)
plt.title('Средний уровень глюкозы по возрастной группе и полу')
plt.xlabel('Возрастная группа (True - люди до 65 лет, False - люди в возрасте 65 лет и старше)')
plt.ylabel('Средний уровень глюкозы')
plt.legend(title='Пол')
plt.show()
```

Средний уровень глюкозы у людей женского пола в возрасте 65 лет и старше выше, чем у остальных.

```python
# ИМТ по возрастной группе и физической активности
plt.figure(figsize=(10, 6))
sns.barplot(x='age_group_Adult', y='BMXBMI', hue='PAQ605', data=df)  # PAQ605_2 - физическая активность
plt.title('Средний ИМТ по возрастной группе и физической активности')
plt.xlabel('Возрастная группа (True - люди до 65 лет, False - люди в возрасте 65 лет и старше)')
plt.ylabel('Средний ИМТ')
plt.legend(title='Физическая активность (1 - Да, 2 - Нет, 7 - Редко)')
plt.show()
```

Физическая активность оказывает влияние на ИМТ, но это влияние различается в зависимости от возрастной группы.
В группе до 65 лет наблюдается тенденция: люди с редкой активностью имеют значительно более низкий ИМТ.
В старшей возрастной группе регулярная физическая активность связана с немного более низким ИМТ.

```python
# Уровень инсулина по полу и физической активности
plt.figure(figsize=(10, 6))
sns.barplot(x='RIAGENDR', y='LBXIN', hue='PAQ605', data=df)
plt.title('Средний уровень инсулина по полу и физической активности')
plt.xlabel('Пол (1 - Мужской, 2 - Женский)')
plt.ylabel('Средний уровень инсулина')
plt.legend(title='Физическая активность (1 - Да, 2 - Нет, 7 - Редко)')
plt.show()
```

Для мужчин:

При регулярной активности средний уровень инсулина около 12
При отсутствии активности уровень немного выше, около 13
При редкой активности уровень значительно ниже, около 4

Для женщин:

При регулярной активности средний уровень инсулина около 11
При отсутствии активности уровень выше, около 12

В целом, уровень инсулина у мужчин и женщин с регулярной активностью и без активности примерно одинаков, с небольшой тенденцией к более высоким значениям у мужчин.
